# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp
FROM master
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
df1 = pd.read_sql_query(QUERY,conn1)

df1['ConvertedComp'].plot.hist(bins=15,alpha=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY=""" SELECT Age 
FROM master"""
df1=pd.read_sql_query(QUERY,conn1)
fig,ax=plt.subplots(figsize=(12,6))
sns.boxplot(x=df1['Age'],ax=ax)
ax.set_xlabel("Age")
ax.set_title("Boxplot")
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY="""SELECT Age,WorkWeekHrs FROM master"""
df1=pd.read_sql_query(QUERY,conn1)
plt.scatter(df1['Age'], df1['WorkWeekHrs'])
plt.xlabel("Age")
plt.ylabel("WorkWeekHrs")
plt.title("Scatter Plot of Age vs WorkWeekHrs")
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY="""SELECT WorkWeekHrs,CodeRevHrs,Age FROM master"""
df1=pd.read_sql_query(QUERY,conn1)
plt.figure(figsize=(12, 8))

# Scatter plot with bubble size based on 'Age'
plt.scatter(df1['WorkWeekHrs'], df1['CodeRevHrs'], s=df1['Age']*10, alpha=0.5)

plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs')
plt.colorbar(label='Age')  # Add a colorbar to show the mapping of bubble sizes to ages
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY=""" SELECT WorkWeekHrs,CodeRevHrs,Age From master"""
df1=pd.read_sql_query(QUERY,conn1)
# Filter the DataFrame for the age group 30 to 35
age_group_df = df1[(df1['Age'] >= 30) & (df1['Age'] <= 35)]

# Calculate the median WorkWeekHrs and CodeRevHrs for the age group
median_work_week_hrs = age_group_df['WorkWeekHrs'].median()
median_code_rev_hrs = age_group_df['CodeRevHrs'].median()

# Create a DataFrame for the stacked chart
data = {
    'Median WorkWeekHrs': [median_work_week_hrs],
    'Median CodeRevHrs': [median_code_rev_hrs]
}
stacked_df = pd.DataFrame(data)

# Plot the stacked chart
stacked_df.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Age Group 30-35')
plt.ylabel('Median Hours')
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xticks([])
plt.legend(title='Hours', loc='upper right')
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY="""SELECT ConvertedComp,Age FROM master"""
df1=pd.read_sql_query(QUERY,conn1)
age_group_df = df1[(df1['Age'] >= 45) & (df1['Age'] <= 60)]

# Calculate the median ConvertedComp for each age
median_converted_comp = age_group_df.groupby('Age')['ConvertedComp'].median()

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(median_converted_comp.index, median_converted_comp.values, marker='o')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY=""" SELECT MainBranch FROM master"""
df1=pd.read_sql_query(QUERY,conn1)
main_branch_counts = df1['MainBranch'].value_counts()

# Plot the horizontal bar chart
plt.figure(figsize=(10, 6))
main_branch_counts.plot(kind='barh', color='skyblue')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.title('Count of MainBranch')
plt.show()
```
